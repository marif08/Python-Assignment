# **# Task 3.2: Basic Groupby Analysis**

```python

# Comprehensive groupby analysis
# Your tasks:

# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

# Your code here:

print('Average Scores:',df.groupby('gender')[['math score','reading score','writing score']].mean())
print('Count of Students:',df.groupby('gender')['gender'].count())

print('Mean Scores:',df.groupby('test preparation course')[['math score','reading score','writing score']].mean())
prep_scores = df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
improvement = prep_scores.loc['completed']-prep_scores.loc['none']
print('Subject that benefits most from test prep:', improvement.idxmax())

print(' Group by test preparation course and Mean scores for each subject:',df.groupby('parental level of education')['math score'].mean())
print('Rank:', df.groupby("parental level of education")["math score"].mean().sort_values(ascending=False))

print('Group by lunch type and Average scores for each subject:',df.groupby('lunch')[['math score','reading score','writing score']].mean())
print('Group by lunch type and Count of students in each lunch category:',df.groupby('lunch')['lunch'].count())

```

```python


# 1. Who performs better in math - males or females?
#    - Calculate average math score by gender
#    - Show the difference between male and female average

print('Average math score by gender:',df.groupby('gender')['math score'].mean())
print('Difference between male and female average:',df.groupby('gender')['math score'].mean()['male']-df.groupby('gender')['math score'].mean()['female'])

# 2. Which subject do students perform best in overall?
#    - Calculate the overall average for math, reading, and writing
#    - Rank the subjects from highest to lowest average score

print('Overall average for math, reading, and writing:',df[['math score','reading score','writing score']].mean())
print('Rank the subjects from highest to lowest average score:',df[['math score','reading score','writing score']].mean().sort_values(ascending=False))

# 3. What's the impact of parental education?
#    - Find the average total score for each parental education level
#    - Identify which education level leads to highest student performance

print('Find the average total score for each parental education level:',df.groupby('parental level of education')[['math score','reading score','writing score']].mean())
print('Identify which education level leads to highest student performance:',df.groupby('parental level of education')[['math score','reading score','writing score']].mean().idxmax())
```
